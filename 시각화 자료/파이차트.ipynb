plt 실행 및 폰트 설정

```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

원하는 파이차트 크기, 모양, value를 넣어서 저장하기

```python
key_word = ['가격','위치','청결','서비스','시설']
values = [596,825,1954,2158,2385]
plt.figure(figsize=(12, 12))
plt.rc('font', size = 30)
w = {"edgecolor": "black", "linewidth":3, "width":0.5}
plt.pie(values, labels=key_word, autopct='%.1f%%', pctdistance=0.8, wedgeprops=w, startangle=90)
plt.title('랜딩관 제주신화월드 호텔앤리조트', size = 50)
plt.savefig('랜딩관 제주신화월드 호텔앤리조트.png')
```
